```python
def find_value(string, regex):
    try:
        return re.findall(regex, string)[0]
    except:
        return None
    
def read_results(filepath):
    results = []
    with open(filepath) as f:
        for i in range(5):
            results.append(float(f.readline()))
    return results
```

```python
df = pd.DataFrame()
for root, dirs, files in os.walk(experiments_path):
    for name in files:
        if name == 'test_results.txt':
            try:
                enemy = find_value(root, r'enemy([\d\,\s\[\]]+)_')
                tournament = find_value(root, r'tournament([A-Za-z]+)')
                mating = find_value(root, r'mating([\d]+)_')
                pop_size = find_value(root, r'pop([\d]+)_')
                patience = find_value(root, r'patience([\d]+)_')
                DPR = find_value(root, r'DPR([\d\.]+)_')
                DRWRP = find_value(root, r'DRWRP([\d\.]+)_')
                mut = find_value(root, r'mut[A-Za-z]+')
                mu = find_value(root, r'mu(\d)sigma')
                sigma = find_value(root, r'sigma(\d)prob')
                mut_prob = find_value(root, r'prob([\d\.]{2,})_') 
                crossover = find_value(root, r'cx[A-Za-z]+')
                results = read_results(os.path.join(root, name))
                df.loc[len(df), ['enemy', 'tournament', 'mating', 'population_size', 'patience', 'DPR', 'DRWRP', 'mut', 'mu','sigma', 'mut_prob', 'crossover', 'test1', 'test2', 'test3', 'test4', 'test5']] = enemy, tournament, mating, pop_size, patience, DPR, DRWRP, mut, mu,sigma, mut_prob, crossover, *results
            except:
                print(os.path.join(root, name))
        else:
            pass # del everythin print(root)
```
